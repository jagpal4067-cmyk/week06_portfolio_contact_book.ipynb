```python
# Week 8 - My Data Story
# This notebook reads a personal text file and analyzes it
# Works in Google Colab

# === Step 1: Upload file from your computer ===
from google.colab import files
import os

print("Please upload your file (assignment txt.txt)")
uploaded = files.upload()

# Rename uploaded file to my_data.txt (assignment requires this name)
filename = list(uploaded.keys())[0]
os.rename(filename, "my_data.txt")
print("File uploaded and renamed to my_data.txt")

# === Step 2: Read and show the file ===
with open("my_data.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

print("\n=== MY DATA STORY ===\n")
for line in lines:
    print(line.strip())

# === Step 3: Calculate some statistics ===
num_lines = len(lines)                                # number of lines
num_words = sum(len(line.split()) for line in lines)  # number of words
longest_line = max(lines, key=len).strip()            # longest line
first_line = lines[0].strip()                         # first line
last_line = lines[-1].strip()                         # last line
total_chars = sum(len(line) for line in lines)        # total characters

# === Step 4: Print results ===
print("\n=== FILE STATISTICS ===")
print(f"Number of lines: {num_lines}")
print(f"Number of words: {num_words}")
print(f"Total characters: {total_chars}")
print(f"Longest line: {longest_line}")
print(f"First line: {first_line}")
print(f"Last line: {last_line}")
```
